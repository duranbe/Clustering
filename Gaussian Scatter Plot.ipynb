```python
import numpy as np
import matplotlib.pyplot as plt
import scipy 
from scipy.spatial import Voronoi, voronoi_plot_2d

from clustering import KMean
```

```python
#https://numpy.org/doc/stable/reference/random/generated/numpy.random.normal.html
mu, sigma = 0,5 # mean and standard deviation
s = np.random.normal(loc=mu,
                     scale=sigma,
                     size=(100,2))
```

### Plot of our random data points

```python
plt.scatter(s[:,0],s[:,1])
```

### Running our KMean algorithm

```python
model = KMean(x=s,k=3,
              distance='euclidean')
```

```python
model.run(n_iter=10)
```

```python
model.middles
```

```python
vor = Voronoi(model.middles)
voronoi_plot_2d(vor)
plt.scatter(s[:,0],s[:,1])
```
